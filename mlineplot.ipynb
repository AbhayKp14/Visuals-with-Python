```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
```

```python
x
```

```python
y = 3*x
```

```python
y
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=5)
plt.title('X vs Y')
plt.xlabel('This is x axis')
plt.ylabel('This is y axis')
plt.show()
```

```python
y1 = 2*x
```

```python
y2 = 3*x
```

```python
plt.plot(x,y1,color='r',linestyle=':')
plt.plot(x,y2,color='y',linewidth=4)
plt.title('Two lines in a plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,color='r',linestyle=':')

plt.subplot(2,1,2)
plt.plot(x,y2,color='y',linewidth=4)
plt.title('Two lines in a plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='r',linestyle=':')

plt.subplot(1,2,2)
plt.plot(x,y2,color='y',linewidth=4)
plt.title('Two lines in a plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.grid(True)
plt.show()
```
