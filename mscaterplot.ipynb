```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=[2,5,9,8,7,4,5,6,3,2]
y=[9,8,7,5,4,2,0,3,6,1]
z=[5,6,8,9,5,2,3,6,5,8]
```

```python
plt.scatter(x,y,marker='*',c='r',s=100)
plt.show()
```

```python
plt.scatter(x,y,marker='*',c='r',s=100)
plt.scatter(x,z,c='g',s=200)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.scatter(x,y,marker='*',c='r',s=100)

plt.subplot(1,2,2)
plt.scatter(x,z,c='g',s=200)
```
