```python
Language = ['Javascript','Python','Html','Css','Ruby','Rstudio']
```

```python
Quantity = [25,45,8,96,7,52]
```

```python
plt.pie(Quantity,labels=Language)
plt.show()
```

```python
plt.pie(Quantity,labels=Language,autopct='%0.1f%%',colors=['yellow','blue','red','grey','pink','violet'])
plt.show()
```

```python
plt.pie(Quantity,labels=Language,radius=2)
plt.pie([100],radius=1,colors=['w'])
plt.show()
```
